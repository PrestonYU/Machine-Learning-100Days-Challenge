# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = './data'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
# 先篩選數值型的欄位
dtype_df = app_train.dtypes.reset_index()
dtype_df.columns = ["Count", "Column Type"]
dtype_df = dtype_df.groupby("Column Type").aggregate('count').reset_index()
dtype_df

numeric_columns = []
for dtype, feature in zip(app_train.dtypes, app_train.columns):
    if dtype == 'float64':
        numeric_columns.append(feature)
    elif dtype == 'int64':
        numeric_columns.append(feature)

print(f'{len(numeric_columns)} Numeric Columns : {numeric_columns}\n')

dtype_select = ['float64','int64']      
numeric_columns1 = list(app_train.columns[list(app_train.dtypes.isin([dtype_select]))])
print(f'{len(numeric_columns1)} Numeric Columns : {numeric_columns1}\n') 
print(numeric_columns == numeric_columns1)  # 此處不清楚為什麼numeric_columns和numeric_columns1結果會不一樣... 用dtype_select沒抓到特徵

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2 ))])
print("%s of remain columns" % len(numeric_columns))

# 檢視這些欄位的數值範圍
for col in numeric_columns:
    app_train.boxplot(col)
    plt.show()
```

```python
# 從上面的圖檢查的結果，至少這三個欄位好像有點可疑

# AMT_INCOME_TOTAL
# REGION_POPULATION_RELATIVE
# OBS_60_CNT_SOCIAL_CIRCLE
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
import statsmodels.api as sm                                      # 載入statsmodels.api

ecdf = sm.distributions.ECDF(app_train['AMT_INCOME_TOTAL'])       # 繪製edcf
plt.plot(np.sort(app_train['AMT_INCOME_TOTAL']), ecdf(np.sort(app_train['AMT_INCOME_TOTAL'])))
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.show()

cdf = app_train['AMT_INCOME_TOTAL'].value_counts().sort_index().cumsum()  # 繪製cdf
plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```

```python
# 最大值落在分布之外
print(app_train['REGION_POPULATION_RELATIVE'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
ecdf = sm.distributions.ECDF(app_train['REGION_POPULATION_RELATIVE'])         # 繪製edcf
plt.plot(np.sort(app_train['REGION_POPULATION_RELATIVE']), ecdf(np.sort(app_train['REGION_POPULATION_RELATIVE'])))
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05])
plt.show()

cdf = app_train['REGION_POPULATION_RELATIVE'].value_counts().sort_index().cumsum()  # 繪製cdf
plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

app_train['REGION_POPULATION_RELATIVE'].hist()
plt.show()

app_train['REGION_POPULATION_RELATIVE'].value_counts()

# 就以這個欄位來說，雖然有資料掉在分布以外，也不算異常，僅代表這間公司在稍微熱鬧的地區有的據點較少，
# 導致 region population relative 在少的部分較為密集，但在大的部分較為疏漏
```

```python
# 最大值落在分布之外
print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
ecdf = sm.distributions.ECDF(app_train['OBS_60_CNT_SOCIAL_CIRCLE'])         # 繪製edcf
plt.plot(np.sort(app_train['OBS_60_CNT_SOCIAL_CIRCLE']), ecdf(np.sort(app_train['OBS_60_CNT_SOCIAL_CIRCLE'])))
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05])
plt.show()


cdf = app_train['OBS_60_CNT_SOCIAL_CIRCLE'].value_counts().sort_index().cumsum()  # 繪製cdf
plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍
plt.show()

app_train['OBS_60_CNT_SOCIAL_CIRCLE'].hist()
plt.show()
print(app_train['OBS_60_CNT_SOCIAL_CIRCLE'].value_counts().sort_index(ascending = False))
```
