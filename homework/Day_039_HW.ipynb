1. LASSO 回歸可以被用來作為 Feature selection 的工具，請了解 LASSO 模型為什麼可用來作 Feature selection

在LASSO回歸下，使用正則化控制模型的複雜程度，因為LASSO可以懲罰權重，可以使得某些特徵為0，並可取出相對不為0，相關度較高的特徵。

2. 當自變數 (X) 存在高度共線性時，Ridge Regression 可以處理這樣的問題嗎?

共線性問題指的是當自變數(X)之間存在較高的線性相關度，若用一般線性回歸將導致迴歸模型的準確性降低。而Ridge Regression即透過放棄最小二乘法的無偏性，以損失部分資訊、降低精準度為標的，減少模型噪度，換得更高的可靠性的回歸方法。
